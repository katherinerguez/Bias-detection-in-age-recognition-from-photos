```python
import matplotlib.pyplot as plt
import numpy as np

# Nombres de los modelos y razas
modelos = ['FairFace Mastrer', 'Modelo 2', 'Modelo 3']
razas = ['White', 'Black', 'Latino_Hispanic', 'East Asian', 'Southeast Asian', 'Indian', 'Middle Eastern']

# Matriz de sesgo (1 = sesgo, 0 = sin sesgo)
# Cada fila corresponde a una raza y cada columna a un modelo
sesgo = np.array([
    [1, 1, 1],  # White
    [1, 1, 1],  # Black
    [1, 0, 1],  # Latino_Hispanic
    [0, 0, 0],  # East Asian
    [0, 1, 1],  # Southeast Asian
    [0, 1, 0],  # Indian
    [1, 1, 1],  # Middle Eastern
])

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Asignar posiciones para los modelos (x) y las razas (y)
x = np.arange(len(modelos))
y = np.arange(len(razas))

# Dibujar los puntos del gráfico solo si hay sesgo (valor 1)
for i in range(len(razas)):
    for j in range(len(modelos)):
        if sesgo[i, j] == 1:
            ax.plot(j, i, 'ro')  # Punto rojo si hay sesgo

# Etiquetas para el eje x e y
ax.set_xticks(x)
ax.set_xticklabels(modelos)
ax.set_yticks(y)
ax.set_yticklabels(razas)

# Título y etiquetas
ax.set_title('Presencia de sesgos por raza en diferentes modelos')
ax.set_xlabel('Modelos')
ax.set_ylabel('Razas')

# Mostrar el gráfico
plt.show()
```
